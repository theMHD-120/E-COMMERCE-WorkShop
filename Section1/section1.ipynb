```python
import numpy as np
import pandas as pd

# ===========================================================
# Code segment 1 (read dataset and simple preprocesses)
# ===========================================================

# Read and clean data
df = pd.read_csv('Housing.csv')
df_cleaned = df.dropna()

# Get numeric columns & preview data
numeric_cols = df_cleaned.select_dtypes(include=['number']).columns
print(df.head(5))
print(df.tail(5))
print(df.describe())

print("==========================================================")

# Convert to numpy array
numeric_array = df_cleaned[numeric_cols].to_numpy()

# Calculate min and max for all columns
min_vals = np.min(numeric_array, axis=0)
max_vals = np.max(numeric_array, axis=0)

# Display results
for i, col in enumerate(numeric_cols):
    print(f"{col}: min = {min_vals[i]}, max = {max_vals[i]}")

# Show non-numeric columns
non_numeric = df_cleaned.select_dtypes(exclude=['number']).columns
if len(non_numeric) > 0:
    print("Non-numeric columns:", list(non_numeric))
```
